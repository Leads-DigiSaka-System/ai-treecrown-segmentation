```python
try:
    import torch
    TORCH_AVAILABLE = True
except ImportError:
    TORCH_AVAILABLE = False

MEM_LIMIT_GB   = 8.0
VRAM_LIMIT_GB  = 12.0
SAM_NATIVE_RES = 1024


def _gpu_available():
    return TORCH_AVAILABLE and torch.cuda.is_available()


def _free_vram_gb():
    if not _gpu_available():
        return 0.0
    props    = torch.cuda.get_device_properties(0)
    reserved = torch.cuda.memory_reserved(0)
    return (props.total_memory - reserved) / 1024**3


def setup_gpu_memory():
    if not _gpu_available():
        print("  No GPU detected — running on CPU")
        return
    os.environ.setdefault('PYTORCH_CUDA_ALLOC_CONF', 'expandable_segments:True')
    total_gb = torch.cuda.get_device_properties(0).total_memory / 1024**3
    frac     = min(VRAM_LIMIT_GB / total_gb, 0.90)
    torch.cuda.set_per_process_memory_fraction(frac)
    print(f"  GPU  : {torch.cuda.get_device_name(0)}")
    print(f"  VRAM : {total_gb:.1f} GB total  →  {frac*100:.0f}% reserved ({frac*total_gb:.1f} GB)")


# ===========================================================================
# STATE
# ===========================================================================

class DetectionState:
    def __init__(self, image_path: str, model, config: dict):
        if not os.path.exists(image_path):
            raise FileNotFoundError(f"Image not found: {image_path}")
        self.image_path = image_path
        self.model      = model
        self.config     = config

        self.original_dims = None
        self.scaled_dims   = None
        self.transform     = None
        self.crs           = None

        self.boxes_list           = []
        self.segmentation_results = []
        self.failed_boxes         = []
        self.gdf                  = None
        self.detection_info       = []

        self.sam2_model        = None
        self._full_image_cache = None
        self._cache_attempted  = False

        self.start_time     = time.time()
        self.start_cpu_time = time.process_time()
        self.stage_times    = {}
        self.stage_memory   = {}
        self.process        = psutil.Process()

    def _maybe_cache_full_image(self):
        if self._cache_attempted:
            return
        self._cache_attempted = True
        if self.original_dims is None:
            return
        orig_w, orig_h = self.original_dims
        image_gb = orig_w * orig_h * 3 / 1024**3
        if image_gb > MEM_LIMIT_GB:
            print(f"  Image {image_gb:.1f} GB > limit {MEM_LIMIT_GB} GB — windowed reads only")
            return
        print(f"  Caching full image ({image_gb:.1f} GB) into RAM…")
        try:
            with rasterio.open(self.image_path) as src:
                if src.count >= 3:
                    data = src.read([1, 2, 3])
                    data = np.transpose(data, (1, 2, 0))
                else:
                    data = src.read(1)
                    data = cv2.cvtColor(data, cv2.COLOR_GRAY2RGB)
            self._full_image_cache = np.clip(data, 0, 255).astype(np.uint8)
            print(f"  ✓ Cached ({self._full_image_cache.nbytes/1024**3:.2f} GB)")
        except MemoryError:
            print("  MemoryError — windowed reads only")
            self._full_image_cache = None

    def read_crop(self, x1: int, y1: int, x2: int, y2: int) -> np.ndarray:
        if self.original_dims is None:
            raise RuntimeError("read_crop called before metadata loaded")
        orig_w, orig_h = self.original_dims
        x1 = max(0, min(x1, orig_w));  y1 = max(0, min(y1, orig_h))
        x2 = max(x1, min(x2, orig_w)); y2 = max(y1, min(y2, orig_h))
        if x2 <= x1 or y2 <= y1:
            return np.zeros((0, 0, 3), dtype=np.uint8)
        if self._full_image_cache is not None:
            return self._full_image_cache[y1:y2, x1:x2].copy()
        window = Window(x1, y1, x2 - x1, y2 - y1)
        try:
            with rasterio.open(self.image_path) as src:
                if src.count >= 3:
                    data = src.read([1, 2, 3], window=window)
                    data = np.transpose(data, (1, 2, 0))
                else:
                    data = src.read(1, window=window)
                    data = cv2.cvtColor(data, cv2.COLOR_GRAY2RGB)
            return np.clip(data, 0, 255).astype(np.uint8)
        except Exception as e:
            print(f"  ⚠ read_crop failed ({e})")
            return np.zeros((y2 - y1, x2 - x1, 3), dtype=np.uint8)

    def log_performance(self, stage_name: str):
        mem     = self.process.memory_info()
        proc_mb = mem.rss / 1024**2
        sys     = psutil.virtual_memory()
        wall    = time.time()         - self.start_time
        cpu     = time.process_time() - self.start_cpu_time
        self.stage_memory[stage_name] = {
            'process_ram_mb':          proc_mb,
            'system_ram_gb':           sys.used      / 1024**3,
            'system_ram_percent':      sys.percent,
            'system_ram_available_gb': sys.available / 1024**3,
            'cpu_percent':             self.process.cpu_percent(interval=0.1),
        }
        self.stage_times[stage_name] = {
            'wall_time': wall, 'cpu_time': cpu,
            'timestamp': datetime.now().strftime('%H:%M:%S'),
        }
        vram_str = ''
        if _gpu_available():
            used = torch.cuda.memory_allocated(0) / 1024**3
            res  = torch.cuda.memory_reserved(0)  / 1024**3
            vram_str = f"\n   VRAM        : {used:.2f} GB alloc / {res:.2f} GB reserved"
        print(f"\n📊 {stage_name}")
        print(f"   Process RAM : {proc_mb:.0f} MB ({proc_mb/1024:.2f} GB)")
        print(f"   System RAM  : {sys.used/1024**3:.2f} GB ({sys.percent:.1f}% used)"
              f" — {sys.available/1024**3:.2f} GB free")
        print(f"   Time        : {wall:.1f}s wall | {cpu:.1f}s CPU" + vram_str)

    def print_performance_summary(self):
        print("\n" + "="*80)
        print(" "*30 + "PERFORMANCE SUMMARY")
        print("="*80)
        print(f"\n⏱️  TIMING:")
        print(f"{'Stage':<42} {'Wall':>10}  {'CPU':>10}")
        print("-"*80)
        for s, t in self.stage_times.items():
            print(f"{s:<42} {t['wall_time']:>8.2f}s  {t['cpu_time']:>8.2f}s")
        tw = time.time()         - self.start_time
        tc = time.process_time() - self.start_cpu_time
        print("-"*80)
        print(f"{'TOTAL':<42} {tw:>8.2f}s  {tc:>8.2f}s")
        print(f"\n💾 MEMORY:")
        print(f"{'Stage':<42} {'Proc RAM':>10}  {'Sys RAM':>10}")
        print("-"*80)
        for s, m in self.stage_memory.items():
            print(f"{s:<42} {m['process_ram_mb']/1024:>8.2f} GB"
                  f"  {m['system_ram_gb']:>6.2f} GB ({m['system_ram_percent']:.0f}%)")
        if self.stage_memory:
            pp = max(m['process_ram_mb'] for m in self.stage_memory.values()) / 1024
            ps = max(m['system_ram_gb']  for m in self.stage_memory.values())
            print("-"*80)
            print(f"{'PEAK':<42} {pp:>8.2f} GB  {ps:>6.2f} GB")
            print(f"\n📈 SUMMARY:")
            print(f"   Total time       : {tw/60:.1f} min")
            print(f"   Peak process RAM : {pp:.2f} GB")
            print(f"   YOLO detections  : {len(self.boxes_list)}")
            n_fb = sum(1 for r in self.segmentation_results if r.get('is_fallback'))
            print(f"   SAM masks        : {len(self.segmentation_results) - n_fb}")
            print(f"   Fallback boxes   : {n_fb}")
        print("="*80)

    def __repr__(self):
        return (f"DetectionState(boxes={len(self.boxes_list)}, "
                f"segs={len(self.segmentation_results)})")


# ===========================================================================
# TRANSITIONS
# ===========================================================================

class StateTransition:
    def __call__(self, state): raise NotImplementedError


class LoadMetadataTransition(StateTransition):
    def __call__(self, state):
        print("Loading image metadata…")
        state.log_performance("1. Metadata")
        try:
            with rasterio.open(state.image_path) as src:
                state.original_dims = (src.width, src.height)
                state.transform     = src.transform
                state.crs           = src.crs
                nb                  = src.count
            print(f"  ✓ {state.original_dims[0]}×{state.original_dims[1]} px  "
                  f"bands={nb}  "
                  f"({state.original_dims[0]*state.original_dims[1]*3/1024**3:.2f} GB raw RGB)")
        except Exception:
            from PIL import Image
            with Image.open(state.image_path) as img:
                w, h = img.size
            state.original_dims = (w, h)
            state.transform     = from_bounds(0, 0, w, h, w, h)
            state.crs           = None
            print(f"  ✓ {w}×{h} px  (plain image — no embedded georeference)")
        state._maybe_cache_full_image()
        return state


class CalculateScaledDimsTransition(StateTransition):
    def __call__(self, state):
        r    = state.config['resolution']
        w, h = state.original_dims
        state.scaled_dims = (int(w * r), int(h * r))
        print(f"  ✓ Scaled dims: {state.scaled_dims[0]}×{state.scaled_dims[1]}  (resolution={r})")
        return state


class YOLODirectInferenceTransition(StateTransition):
    """
    Feed the full image to YOLO in a single forward pass — no tiling.
    The image path is passed directly to model.predict() so YOLO handles
    its own internal resizing, exactly like a plain inference script would.
    Bounding boxes and masks are read back from the result object and stored
    on state for the SAM2 segmentation stage.
    """

    def __call__(self, state):
        state.log_performance("2. YOLO Detection")
        cfg     = state.config
        conf    = cfg['conf_threshold']
        iou     = cfg['iou_threshold']
        max_det = cfg['max_det']
        imgsz   = cfg.get('yolo_imgsz', 640)
        device  = 'cuda' if _gpu_available() else 'cpu'

        orig_w, orig_h = state.original_dims
        print(f"  Full-image inference  imgsz={imgsz}  "
              f"orig={orig_w}×{orig_h}  device={device}")

        # Pass the image path directly — identical to a standalone predict call
        results = state.model.predict(
            source=state.image_path,
            conf=conf,
            iou=iou,
            imgsz=imgsz,
            max_det=max_det,
            verbose=False,
            device=device,
        )

        boxes_list = []
        if results and results[0].boxes is not None:
            boxes = results[0].boxes.xyxy.cpu().numpy()   # (N,4) xyxy in original px
            confs = results[0].boxes.conf.cpu().numpy()

            # Log mask availability
            if results[0].masks is not None:
                print(f"  YOLO masks available: {len(results[0].masks.data)}")

            for i in range(len(boxes)):
                x1, y1, x2, y2 = boxes[i]
                boxes_list.append({
                    'bbox':       [int(x1), int(y1), int(x2), int(y2)],
                    'confidence': float(confs[i]),
                })

        del results
        state.boxes_list = boxes_list
        print(f"  ✓ {len(state.boxes_list)} boxes detected")
        return state


# Keep the tiled version available but not used by default
class YOLOStreamingTransition(StateTransition):
    """
    Tiled YOLO inference for large aerial images.

    Root causes of under-detection in the original and their fixes:
    ───────────────────────────────────────────────────────────────
    1. BUG — double division by `resolution` during bbox mapping.
       Model detections were divided by resolution twice, shifting all
       boxes to the wrong location so some appeared out-of-image.
       FIX: remove resolution from the coord formula entirely; tiles are
       defined in original-image pixel space from the start.

    2. BUG — tile size too small (640 px) relative to actual crown size.
       At 640 px tiles on a high-res image, many crowns occupied only a
       few pixels, falling below the model's effective detection range.
       FIX: tile_size now defines the crop region in the original image
       (default 1280 px). The model still receives yolo_imgsz (640) but
       sees a larger, more context-rich region of the scene.

    3. BUG — overlap too small (128 px = 20 %% of 640).
       Trees near tile edges appeared only partially in each tile.
       FIX: overlap_frac=0.5 (50 %%) ensures every tree appears fully
       inside at least one tile regardless of position.
    """

    def __call__(self, state):
        state.log_performance("2. YOLO Detection")
        cfg        = state.config
        tile_size  = cfg.get('tile_size',    1280)  # crop region in original-image pixels
        overlap_f  = cfg.get('overlap_frac', 0.5)   # overlap as fraction of tile_size
        yolo_imgsz = cfg.get('yolo_imgsz',   640)   # model inference resolution
        conf       = cfg['conf_threshold']
        iou        = cfg['iou_threshold']
        max_det    = cfg['max_det']
        batch_size = cfg.get('batch_size', 8)

        orig_w, orig_h = state.original_dims
        overlap = int(tile_size * overlap_f)
        step    = tile_size - overlap

        # All tile positions are in original-image pixel space
        tile_positions = [
            (x, y,
             min(x + tile_size, orig_w),
             min(y + tile_size, orig_h))
            for y in range(0, orig_h, step)
            for x in range(0, orig_w, step)
        ]

        device = 'cuda' if _gpu_available() else 'cpu'
        print(f"  {len(tile_positions)} tiles | "
              f"crop={tile_size}px  overlap={int(overlap_f*100)}%%  "
              f"model={yolo_imgsz}px  batch={batch_size}  device={device}")

        state.boxes_list = self._run_batches(
            state, tile_positions, yolo_imgsz,
            conf, iou, max_det, batch_size, device, orig_w, orig_h
        )
        print(f"  ✓ {len(state.boxes_list)} raw boxes (before NMS/WBF)")
        return state

    def _run_batches(self, state, tile_positions, yolo_imgsz,
                     conf, iou, max_det, batch_size, device, orig_w, orig_h):
        all_boxes = []

        for start in tqdm(range(0, len(tile_positions), batch_size), desc="YOLO tiles"):
            batch  = tile_positions[start : start + batch_size]
            images, meta = [], []

            for (tx, ty, tx2, ty2) in batch:
                # Read the actual pixel region from the original image
                crop     = state.read_crop(tx, ty, tx2, ty2)
                actual_w = tx2 - tx
                actual_h = ty2 - ty

                # Resize to model resolution — sx/sy convert back to orig pixels
                crop_sq = cv2.resize(
                    crop, (yolo_imgsz, yolo_imgsz),
                    interpolation=cv2.INTER_LINEAR
                )
                sx = actual_w / yolo_imgsz   # model-x → original-x offset
                sy = actual_h / yolo_imgsz   # model-y → original-y offset

                images.append(np.ascontiguousarray(crop_sq))
                meta.append((tx, ty, sx, sy))

            results = state.model.predict(
                source=images,
                conf=conf,
                iou=iou,
                max_det=max_det,
                imgsz=yolo_imgsz,
                verbose=False,
                device=device,
            )

            for result, (tx, ty, sx, sy) in zip(results, meta):
                if result.boxes is None or len(result.boxes) == 0:
                    continue
                boxes = result.boxes.xyxy.cpu().numpy()
                confs = result.boxes.conf.cpu().numpy()
                for i in range(len(boxes)):
                    x1, y1, x2, y2 = boxes[i]
                    # Single correct formula: scale model coords then offset to image space
                    all_boxes.append({
                        'bbox': [
                            int(np.clip(x1 * sx + tx, 0, orig_w)),
                            int(np.clip(y1 * sy + ty, 0, orig_h)),
                            int(np.clip(x2 * sx + tx, 0, orig_w)),
                            int(np.clip(y2 * sy + ty, 0, orig_h)),
                        ],
                        'confidence': float(confs[i]),
                    })

            del images, results

        return all_boxes


class NMSTransition(StateTransition):
    """
    Two-stage deduplication for tiled inference.

    Stage 1 — Weighted Box Fusion (WBF):
        Same tree detected across overlapping tiles → cluster by IoU,
        average coords weighted by confidence into one box.
        This recovers correctly-centred boxes for edge-straddling crowns.

    Stage 2 — Lenient Hard-NMS (default iou_thr=0.50):
        Suppress any remaining true duplicates after WBF.
        Threshold is HIGH intentionally — neighbouring trees that share
        some bbox area must NOT be suppressed. Only boxes overlapping
        by more than 50 %% of their union are treated as duplicates.
    """

    def __call__(self, state):
        state.log_performance("3. NMS")
        if not state.config.get('apply_nms', True) or not state.boxes_list:
            return state

        wbf_iou = state.config.get('wbf_iou', 0.40)
        nms_iou = state.config.get('nms_iou', 0.50)
        before  = len(state.boxes_list)

        fused = self._wbf(state.boxes_list, wbf_iou)
        final = self._hard_nms(fused, nms_iou)

        print(f"  WBF : {before} → {len(fused)}  (fused tile-duplicate detections, IoU≥{wbf_iou})")
        print(f"  NMS : {len(fused)} → {len(final)}  (removed true duplicates, IoU≥{nms_iou})")
        state.boxes_list = final
        return state

    @staticmethod
    def _iou(a, b):
        ix1 = max(a[0], b[0]); iy1 = max(a[1], b[1])
        ix2 = min(a[2], b[2]); iy2 = min(a[3], b[3])
        if ix2 <= ix1 or iy2 <= iy1: return 0.0
        inter = (ix2 - ix1) * (iy2 - iy1)
        union = (a[2]-a[0])*(a[3]-a[1]) + (b[2]-b[0])*(b[3]-b[1]) - inter
        return inter / union if union > 0 else 0.0

    def _wbf(self, boxes, iou_thr):
        """Weighted Box Fusion — merge same-tree detections from overlapping tiles."""
        boxes    = sorted(boxes, key=lambda b: b['confidence'], reverse=True)
        clusters = []
        for box in boxes:
            matched = False
            for cluster in clusters:
                if self._iou(box['bbox'], cluster[0]['bbox']) >= iou_thr:
                    cluster.append(box)
                    matched = True
                    break
            if not matched:
                clusters.append([box])
        fused = []
        for cluster in clusters:
            w     = [b['confidence'] for b in cluster]
            total = sum(w)
            fused.append({
                'bbox': [
                    int(sum(wi * b['bbox'][k] for wi, b in zip(w, cluster)) / total)
                    for k in range(4)
                ],
                'confidence': float(max(b['confidence'] for b in cluster)),
            })
        return fused

    def _hard_nms(self, boxes, iou_thr):
        """Greedy NMS with lenient threshold to protect neighbouring trees."""
        boxes   = sorted(boxes, key=lambda b: b['confidence'], reverse=True)
        keep    = []
        removed = set()
        for i, bi in enumerate(boxes):
            if i in removed: continue
            keep.append(bi)
            for j in range(i + 1, len(boxes)):
                if j in removed: continue
                if self._iou(bi['bbox'], boxes[j]['bbox']) > iou_thr:
                    removed.add(j)
        return keep


class SAM2InitTransition(StateTransition):
    def __call__(self, state):
        state.log_performance("4. SAM2 Init")
        ckpt = state.config['sam_checkpoint']
        if not os.path.exists(ckpt):
            raise FileNotFoundError(f"SAM 2.1 checkpoint not found: {ckpt}")
        from ultralytics import SAM
        state.sam2_model = SAM(ckpt)
        print("  ✓ SAM 2.1 loaded")
        return state


# ===========================================================================
# SAM2 — v5: re-queue out-of-crop boxes as solo inferences
# ===========================================================================

class SAM2StreamingTransition(StateTransition):

    def __call__(self, state):
        print(f"  Segmenting {len(state.boxes_list)} boxes…")
        state.segmentation_results, state.failed_boxes = self._segment(state)
        n_ok = len(state.segmentation_results)
        n_fb = sum(1 for r in state.segmentation_results if r.get('is_fallback'))
        print(f"  ✓ {n_ok} results  ({n_ok-n_fb} masks  {n_fb} fallbacks"
              f"  {len(state.failed_boxes)} hard-failures)")
        state.log_performance("5. SAM2 Segmentation")
        return state

    # ------------------------------------------------------------------
    def _segment(self, state):
        results_dict = {}
        failed       = []
        retry_queue  = []   # (orig_index, box_info) — processed solo after main loop

        cfg         = state.config
        base_batch  = cfg.get('sam2_batch_size', 4)
        pad         = cfg.get('sam2_crop_pad', 80)
        max_crop_px = cfg.get('sam2_max_crop', 1024)
        min_crop_px = cfg.get('sam2_min_crop', 64)
        grid_cell   = cfg.get('sam2_sort_grid', 800)
        device      = 'cuda' if _gpu_available() else 'cpu'
        orig_w, orig_h = state.original_dims

        counters = dict(empty_mask=0, no_contour=0, degen_poly=0,
                        sam_fail=0, oom=0, requeued=0, solo_ok=0, solo_fail=0)

        # ── Spatial sort ──────────────────────────────────────────────
        indexed = sorted(
            enumerate(state.boxes_list),
            key=lambda t: (
                t[1]['bbox'][1] // grid_cell,
                t[1]['bbox'][0] // grid_cell,
            )
        )

        pbar = tqdm(total=len(indexed), desc="SAM2")
        batch_start = 0

        while batch_start < len(indexed):
            if _gpu_available():
                free_gb   = _free_vram_gb()
                dyn_batch = max(1, min(base_batch, int(base_batch * free_gb / 2.0)))
            else:
                dyn_batch = base_batch

            chunk        = indexed[batch_start : batch_start + dyn_batch]
            orig_indices = [t[0] for t in chunk]
            batch_boxes  = [t[1] for t in chunk]
            batch_start += len(chunk)

            # ── Tight padded crop ──────────────────────────────────────
            all_x = ([b['bbox'][0] for b in batch_boxes] +
                     [b['bbox'][2] for b in batch_boxes])
            all_y = ([b['bbox'][1] for b in batch_boxes] +
                     [b['bbox'][3] for b in batch_boxes])
            cx1 = max(0,      min(all_x) - pad)
            cy1 = max(0,      min(all_y) - pad)
            cx2 = min(orig_w, max(all_x) + pad)
            cy2 = min(orig_h, max(all_y) + pad)

            # Hard-cap — record original crop extent BEFORE capping
            uncapped_cx1, uncapped_cy1 = cx1, cy1
            uncapped_cx2, uncapped_cy2 = cx2, cy2

            if (cx2 - cx1) > max_crop_px:
                mid = (cx1 + cx2) // 2
                cx1 = max(0,      mid - max_crop_px // 2)
                cx2 = min(orig_w, mid + max_crop_px // 2)
            if (cy2 - cy1) > max_crop_px:
                mid = (cy1 + cy2) // 2
                cy1 = max(0,      mid - max_crop_px // 2)
                cy2 = min(orig_h, mid + max_crop_px // 2)

            # Minimum crop size
            if (cx2 - cx1) < min_crop_px:
                mid = (cx1 + cx2) // 2
                cx1 = max(0,      mid - min_crop_px // 2)
                cx2 = min(orig_w, mid + min_crop_px // 2)
            if (cy2 - cy1) < min_crop_px:
                mid = (cy1 + cy2) // 2
                cy1 = max(0,      mid - min_crop_px // 2)
                cy2 = min(orig_h, mid + min_crop_px // 2)

            # ── KEY FIX: detect boxes outside the capped crop ─────────
            # Instead of silently marking them degenerate, re-queue them.
            in_crop  = []
            out_crop = []
            for idx, b in zip(orig_indices, batch_boxes):
                bx1, by1, bx2, by2 = b['bbox']
                # Box centre must be inside the crop; partial overlap is fine
                bx_centre = (bx1 + bx2) / 2
                by_centre = (by1 + by2) / 2
                if cx1 <= bx_centre <= cx2 and cy1 <= by_centre <= cy2:
                    in_crop.append((idx, b))
                else:
                    out_crop.append((idx, b))

            # Re-queue out-of-crop boxes for solo processing
            for idx, b in out_crop:
                retry_queue.append((idx, b))
                counters['requeued'] += 1

            orig_indices = [t[0] for t in in_crop]
            batch_boxes  = [t[1] for t in in_crop]

            if not batch_boxes:
                pbar.update(len(chunk)); continue

            crop_raw = state.read_crop(cx1, cy1, cx2, cy2)
            if crop_raw.size == 0:
                for idx, b in zip(orig_indices, batch_boxes):
                    retry_queue.append((idx, b))
                    counters['requeued'] += 1
                pbar.update(len(chunk)); continue

            result = self._run_sam_on_crop(
                state, crop_raw, cx1, cy1, orig_indices, batch_boxes,
                max_crop_px, device, counters
            )
            results_dict.update(result)
            pbar.update(len(chunk))

        pbar.close()

        # ── Solo retry pass ───────────────────────────────────────────
        if retry_queue:
            print(f"\n  Re-processing {len(retry_queue)} out-of-crop boxes solo…")
            solo_pbar = tqdm(total=len(retry_queue), desc="SAM2 retry")
            for idx, b in retry_queue:
                bx1, by1, bx2, by2 = b['bbox']
                # Build a crop centred on this single box
                bw = bx2 - bx1; bh = by2 - by1
                solo_pad = max(pad, max(bw, bh) // 2)
                sx1 = max(0,      bx1 - solo_pad)
                sy1 = max(0,      by1 - solo_pad)
                sx2 = min(orig_w, bx2 + solo_pad)
                sy2 = min(orig_h, by2 + solo_pad)
                # Cap to SAM native res
                if (sx2 - sx1) > max_crop_px:
                    mid = (sx1 + sx2) // 2
                    sx1 = max(0,      mid - max_crop_px // 2)
                    sx2 = min(orig_w, mid + max_crop_px // 2)
                if (sy2 - sy1) > max_crop_px:
                    mid = (sy1 + sy2) // 2
                    sy1 = max(0,      mid - max_crop_px // 2)
                    sy2 = min(orig_h, mid + max_crop_px // 2)

                crop_raw = state.read_crop(sx1, sy1, sx2, sy2)
                if crop_raw.size == 0:
                    results_dict[idx] = self._fb(b)
                    counters['solo_fail'] += 1
                    solo_pbar.update(1); continue

                result = self._run_sam_on_crop(
                    state, crop_raw, sx1, sy1, [idx], [b],
                    max_crop_px, device, counters
                )
                if result.get(idx, {}).get('is_fallback', True):
                    counters['solo_fail'] += 1
                else:
                    counters['solo_ok'] += 1
                results_dict.update(result)
                solo_pbar.update(1)
            solo_pbar.close()

        print(f"\n  Failure breakdown:")
        print(f"    Requeued (out-of-crop) : {counters['requeued']}")
        print(f"    Solo retry — OK        : {counters['solo_ok']}")
        print(f"    Solo retry — fail      : {counters['solo_fail']}")
        print(f"    Empty mask             : {counters['empty_mask']}")
        print(f"    No contour             : {counters['no_contour']}")
        print(f"    Degen polygon          : {counters['degen_poly']}")
        print(f"    SAM fail               : {counters['sam_fail']}")
        print(f"    OOM events             : {counters['oom']}")

        results = [
            results_dict.get(i, self._fb(state.boxes_list[i]))
            for i in range(len(state.boxes_list))
        ]
        return results, failed

    # ------------------------------------------------------------------
    def _run_sam_on_crop(self, state, crop_raw, cx1, cy1,
                          orig_indices, batch_boxes,
                          max_crop_px, device, counters):
        """
        Run SAM on a single crop for a set of boxes.
        Returns a dict of {orig_index: result}.
        """
        results = {}
        raw_h, raw_w = crop_raw.shape[:2]
        orig_w, orig_h = state.original_dims

        # Resize crop to SAM native resolution
        scale_to_sam = min(SAM_NATIVE_RES / raw_w, SAM_NATIVE_RES / raw_h, 1.0)
        if scale_to_sam < 1.0:
            sam_w = int(raw_w * scale_to_sam)
            sam_h = int(raw_h * scale_to_sam)
            crop_sam = cv2.resize(crop_raw, (sam_w, sam_h), interpolation=cv2.INTER_LINEAR)
        else:
            sam_w, sam_h = raw_w, raw_h
            scale_to_sam = 1.0
            crop_sam = crop_raw

        # Remap bbox coords into SAM crop space
        bboxes_sam = np.array([
            [
                max(0,     (b['bbox'][0] - cx1) * scale_to_sam),
                max(0,     (b['bbox'][1] - cy1) * scale_to_sam),
                min(sam_w, (b['bbox'][2] - cx1) * scale_to_sam),
                min(sam_h, (b['bbox'][3] - cy1) * scale_to_sam),
            ]
            for b in batch_boxes
        ], dtype=np.float32)

        # Degenerate box check — these are now genuine failures, not crop issues
        valid = ((bboxes_sam[:, 2] > bboxes_sam[:, 0] + 2) &
                 (bboxes_sam[:, 3] > bboxes_sam[:, 1] + 2))
        for idx, b, v in zip(orig_indices, batch_boxes, valid):
            if not v:
                results[idx] = self._fb(b)
        orig_indices = [i for i, v in zip(orig_indices, valid) if v]
        batch_boxes  = [b for b, v in zip(batch_boxes,  valid) if v]
        bboxes_sam   = bboxes_sam[valid]

        if not batch_boxes:
            return results

        # OOM-resilient inference
        sam_out   = None
        act_idx   = list(orig_indices)
        act_boxes = list(batch_boxes)
        act_bb    = bboxes_sam

        for attempt in range(3):
            try:
                if _gpu_available():
                    torch.cuda.empty_cache()
                sam_out = state.sam2_model(
                    crop_sam, bboxes=act_bb, verbose=False, device=device
                )
                break
            except RuntimeError as e:
                if 'out of memory' not in str(e).lower() or attempt == 2:
                    counters['sam_fail'] += len(act_boxes)
                    sam_out = None; break
                counters['oom'] += 1
                if _gpu_available(): torch.cuda.empty_cache()
                gc.collect()
                half = max(1, len(act_boxes) // 2)
                for idx, b in zip(act_idx[half:], act_boxes[half:]):
                    results[idx] = self._fb(b)
                act_idx   = act_idx[:half]
                act_boxes = act_boxes[:half]
                act_bb    = act_bb[:half]
                if not act_boxes:
                    sam_out = None; break

        if sam_out is None or not sam_out or sam_out[0].masks is None:
            for idx, b in zip(act_idx, act_boxes):
                results[idx] = self._fb(b)
                counters['sam_fail'] += 1
            return results

        # Parse masks → polygons
        masks = sam_out[0].masks.data.cpu().numpy()
        if masks.ndim == 2:
            masks = masks[np.newaxis]

        for i, (idx, box_info) in enumerate(zip(act_idx, act_boxes)):
            if i >= masks.shape[0]:
                results[idx] = self._fb(box_info); continue

            mask = masks[i]
            if mask.shape != (raw_h, raw_w):
                mask = cv2.resize(
                    mask.astype(np.float32), (raw_w, raw_h),
                    interpolation=cv2.INTER_LINEAR
                )
            binary = (mask > 0.5).astype(np.uint8)

            if binary.sum() < 10:
                counters['empty_mask'] += 1
                results[idx] = self._fb(box_info); continue

            contours, _ = cv2.findContours(
                binary, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE
            )
            if not contours:
                counters['no_contour'] += 1
                results[idx] = self._fb(box_info); continue

            contour  = max(contours, key=cv2.contourArea)
            arc      = cv2.arcLength(contour, True)
            bbox     = box_info['bbox']
            box_w    = bbox[2] - bbox[0]; box_h = bbox[3] - bbox[1]
            eps_frac = 0.004 if min(box_w, box_h) < 30 else 0.002
            approx   = cv2.approxPolyDP(contour, eps_frac * arc, True)

            if len(approx) < 3:
                counters['degen_poly'] += 1
                results[idx] = self._fb(box_info); continue

            coords    = approx.reshape(-1, 2) + np.array([cx1, cy1])
            ys, xs    = np.where(binary > 0)
            iou_score = 0.0
            if len(ys):
                ix1 = max(int(xs.min()) + cx1, bbox[0])
                iy1 = max(int(ys.min()) + cy1, bbox[1])
                ix2 = min(int(xs.max()) + cx1, bbox[2])
                iy2 = min(int(ys.max()) + cy1, bbox[3])
                if ix2 > ix1 and iy2 > iy1:
                    inter     = (ix2-ix1)*(iy2-iy1)
                    box_area  = (bbox[2]-bbox[0])*(bbox[3]-bbox[1])
                    mask_area = int(len(ys))
                    union     = box_area + mask_area - inter
                    iou_score = inter/union if union > 0 else 0.0

            results[idx] = {
                'coords':      coords,
                'bbox':        bbox,
                'confidence':  box_info['confidence'],
                'sam_score':   float(iou_score),
                'is_fallback': False,
            }

        del sam_out
        return results

    @staticmethod
    def _fb(box_info):
        bx1, by1, bx2, by2 = box_info['bbox']
        return {
            'coords':      np.array([[bx1,by1],[bx2,by1],[bx2,by2],[bx1,by2]]),
            'bbox':        box_info['bbox'],
            'confidence':  box_info['confidence'],
            'sam_score':   0.0,
            'is_fallback': True,
        }


# ===========================================================================
# GeoDataFrame, Visualization, Save — unchanged
# ===========================================================================

class CreateGeoDataFrameTransition(StateTransition):
    def __call__(self, state):
        print(f"  Building GeoDataFrame ({len(state.segmentation_results)} results)…")
        state.log_performance("6. GeoDataFrame")
        if not state.segmentation_results:
            state.gdf = gpd.GeoDataFrame(); state.detection_info = []; return state
        polygons, info = [], []
        for r in tqdm(state.segmentation_results, desc="Polygons"):
            try:
                poly = Polygon(r['coords'])
                if not poly.is_valid: poly = poly.buffer(0)
                area_px = poly.area
            except Exception: continue
            geo_coords = [rasterio.transform.xy(state.transform, py, px)
                          for px, py in r['coords']]
            try:
                geo = Polygon(geo_coords)
                if not geo.is_valid: geo = geo.buffer(0)
            except Exception: continue
            polygons.append(geo)
            info.append({
                'area_pixels': area_px,
                'confidence':  r['confidence'],
                'sam_score':   r.get('sam_score', 0),
                'is_fallback': r.get('is_fallback', False),
                'num_points':  len(r['coords']),
            })
        if not polygons:
            state.gdf = gpd.GeoDataFrame(); state.detection_info = []; return state
        state.gdf = gpd.GeoDataFrame({
            'geometry':    polygons,
            'area_pixels': [d['area_pixels'] for d in info],
            'confidence':  [d['confidence']  for d in info],
            'sam_score':   [d['sam_score']   for d in info],
            'is_fallback': [d['is_fallback'] for d in info],
            'num_points':  [d['num_points']  for d in info],
        }, crs=state.crs)
        state.detection_info = info
        print(f"  ✓ {len(state.gdf)} features")
        return state


class VisualizationTransition(StateTransition):
    def __call__(self, state):
        state.log_performance("7. Visualization")
        cfg        = state.config
        output_dir = cfg['output_dir']
        name       = cfg.get('image_name', 'output')
        max_dim    = cfg.get('max_viz_dimension', 4000)
        strip_src  = cfg.get('viz_strip_height', 512)
        orig_w, orig_h = state.original_dims
        scale  = min(max_dim / orig_w, max_dim / orig_h, 1.0)
        out_w  = int(orig_w * scale)
        out_h  = int(orig_h * scale)
        os.makedirs(output_dir, exist_ok=True)
        output_path = os.path.join(output_dir, f"{name}_segmentation.png")

        scaled = []
        for r in state.segmentation_results:
            sc = (r['coords'] * scale).astype(np.int32)
            sc[:, 0] = np.clip(sc[:, 0], 0, out_w - 1)
            sc[:, 1] = np.clip(sc[:, 1], 0, out_h - 1)
            scaled.append({**r, '_sc': sc})

        strip_out_h = max(1, int(strip_src * scale))
        bucket: dict[int, list] = defaultdict(list)
        for r in scaled:
            y_min = int(r['_sc'][:, 1].min())
            y_max = int(r['_sc'][:, 1].max())
            for si in range(y_min // strip_out_h, y_max // strip_out_h + 1):
                bucket[si].append(r)

        canvas = np.zeros((out_h, out_w, 3), dtype=np.uint8)

        for si in tqdm(range(0, out_h, strip_out_h), desc="Viz strips"):
            y0 = si; y1 = min(si + strip_out_h, out_h)
            src_y0 = int(y0 / scale); src_y1 = int(y1 / scale)
            src = state.read_crop(0, src_y0, orig_w, src_y1)
            if src.size == 0: continue
            strip = cv2.resize(src, (out_w, y1 - y0), interpolation=cv2.INTER_AREA)

            for r in bucket.get(si // strip_out_h, []):
                coords = r['_sc'].copy()
                coords[:, 1] -= y0
                if r.get('is_fallback'):
                    cv2.polylines(strip, [coords], True, (255, 140, 0), 2, cv2.LINE_AA)
                else:
                    # Red outline only — no fill, just like the reference photo
                    cv2.polylines(strip, [coords], True, (40,  0,   0), 4, cv2.LINE_AA)  # dark shadow
                    cv2.polylines(strip, [coords], True, (210, 30,  30), 2, cv2.LINE_AA) # red
                    cv2.polylines(strip, [coords], True, (255, 90,  60), 1, cv2.LINE_AA) # bright inner

            canvas[y0:y1] = strip
            del src, strip

        # ── overlay ──────────────────────────────────────────────────────────
        n     = len(state.segmentation_results)
        avg_c = np.mean([r['confidence'] for r in state.segmentation_results]) if n else 0
        avg_s = np.mean([r['sam_score']  for r in state.segmentation_results]) if n else 0
        n_fb  = sum(1 for r in state.segmentation_results if r.get('is_fallback'))

        lines = [
            "TREE CROWN SEGMENTATION",
            f"Trees Detected: {n:,}",
            f"Avg Confidence: {avg_c:.3f}   Avg SAM Score: {avg_s:.3f}",
            f"SAM Masks: {n - n_fb:,}   Fallbacks: {n_fb:,}",
        ]
        font_sizes = [20, 13, 11, 11]
        pad        = 14
        line_gap   = 6

        from PIL import ImageFont, ImageDraw, Image as PILImage

        def load_arial(size):
            for path in [
                "/usr/share/fonts/truetype/msttcorefonts/Arial.ttf",
                "/usr/share/fonts/truetype/liberation/LiberationSans-Regular.ttf",
                "/usr/share/fonts/liberation/LiberationSans-Regular.ttf",
                "/usr/share/fonts/truetype/dejavu/DejaVuSans.ttf",
            ]:
                if os.path.exists(path):
                    return ImageFont.truetype(path, size)
            return ImageFont.load_default()

        fonts   = [load_arial(s) for s in font_sizes]
        dummy   = ImageDraw.Draw(PILImage.new("RGB", (1, 1)))
        widths  = [dummy.textlength(l, font=f) for l, f in zip(lines, fonts)]
        heights = [dummy.textbbox((0, 0), l, font=f)[3] for l, f in zip(lines, fonts)]

        box_w = int(max(widths))  + pad * 2
        box_h = int(sum(heights)) + line_gap * (len(lines) - 1) + pad * 2

        pil_img = PILImage.fromarray(canvas)
        draw    = ImageDraw.Draw(pil_img, "RGBA")

        x0, y0 = 20, 20
        x1, y1 = x0 + box_w, y0 + box_h

        # White box with subtle grey border
        draw.rectangle([x0, y0, x1, y1], fill=(255, 255, 255, 245))
        draw.rectangle([x0, y0, x1, y1], outline=(180, 180, 180, 255), width=1)

        cy = y0 + pad
        for line, font, h in zip(lines, fonts, heights):
            draw.text((x0 + pad, cy), line, font=font, fill=(30, 30, 30))
            cy += h + line_gap

        canvas = np.array(pil_img)

        cv2.imwrite(output_path, cv2.cvtColor(canvas, cv2.COLOR_RGB2BGR),
                    [cv2.IMWRITE_PNG_COMPRESSION, 6])
        print(f"  ✓ {output_path}  ({os.path.getsize(output_path)/1024**2:.1f} MB)")
        del canvas, scaled; gc.collect()
        return state


class SaveResultsTransition(StateTransition):
    def __call__(self, state):
        state.log_performance("8. Save Results")
        if state.gdf is None or len(state.gdf) == 0:
            print("  No detections to save"); return state
        output_dir = state.config['output_dir']
        name       = state.config.get('image_name', 'output')
        os.makedirs(output_dir, exist_ok=True)
        p = os.path.join(output_dir, f"{name}_detections.geojson")
        try:
            state.gdf.to_file(p, driver='GeoJSON')
            crs_label = str(state.gdf.crs) if state.gdf.crs is not None else "pixel coords (no CRS)"
            print(f"  ✓ GeoJSON: {p}  [{crs_label}]")
        except Exception as e:
            print(f"  ⚠ GeoJSON: {e}")
        p = os.path.join(output_dir, f"{name}_detections.csv")
        try:   state.gdf.drop(columns='geometry').to_csv(p, index=False); print(f"  ✓ CSV: {p}")
        except Exception as e: print(f"  ⚠ CSV: {e}")
        if state.detection_info:
            stats = {
                'total_detections':        len(state.gdf),
                'successful_sam_masks':    sum(1 for d in state.detection_info if d['sam_score'] > 0),
                'fallback_boxes':          sum(1 for d in state.detection_info if d['sam_score'] == 0),
                'average_area_pixels':     float(np.mean([d['area_pixels'] for d in state.detection_info])),
                'median_area_pixels':      float(np.median([d['area_pixels'] for d in state.detection_info])),
                'average_yolo_confidence': float(np.mean([d['confidence']   for d in state.detection_info])),
                'average_sam_score':       float(np.mean([d['sam_score']    for d in state.detection_info])),
            }
            p = os.path.join(output_dir, f"{name}_statistics.json")
            try:
                with open(p, 'w') as f: json.dump(stats, f, indent=2)
                print(f"  ✓ Statistics: {p}")
            except Exception as e: print(f"  ⚠ Statistics: {e}")
        print(f"  ✓ Done in {(time.time()-state.start_time)/60:.1f} min")
        return state


# ===========================================================================
# PIPELINE
# ===========================================================================

class DetectionPipeline:
    def __init__(self, transitions): self.transitions = transitions

    def run(self, state):
        for i, t in enumerate(self.transitions):
            print(f"\n{'='*60}")
            print(f"[{i+1}/{len(self.transitions)}] {t.__class__.__name__}")
            print("="*60)
            state = t(state)
        return state


def create_detection_pipeline(config):
    """
    Direct (full-image) inference pipeline for basemap screenshots.

    The entire screenshot is fed to YOLO in a single forward pass — no tiling,
    no boundary splits, no missed crowns. NMSTransition is intentionally omitted
    as it was only needed to de-duplicate boxes from overlapping tiles.
    """
    return DetectionPipeline([
        LoadMetadataTransition(),
        CalculateScaledDimsTransition(),
        YOLODirectInferenceTransition(),   # full-image inference — no tiling, no NMS
        SAM2InitTransition(),
        SAM2StreamingTransition(),
        CreateGeoDataFrameTransition(),
        VisualizationTransition(),
        SaveResultsTransition(),
    ])


# ===========================================================================
# ENTRY POINT
# ===========================================================================

if __name__ == "__main__":
    setup_gpu_memory()

    # =========================================================================
    # INPUT — Screenshot from basemap (Google Maps, etc.)
    # The image path is passed directly to YOLO — no tiling, no preprocessing.
    # Coordinates are preserved if the image has embedded georeferencing;
    # otherwise pixel coordinates are used and GeoJSON is still written.
    # =========================================================================

    SCREENSHOT_PATH = (
        '/content/drive/MyDrive/AGRI/TreeCrown_Segmentation/test/Screenshot 2026-04-22 064955.png'
    )

    config = {
        # ── Direct inference — mirrors standalone predict call ─────────────────
        # yolo_imgsz : passed directly to model.predict(); YOLO handles resizing.
        #              640 = model native res; 1280 recovers more small crowns.
        # conf/iou   : same thresholds as the standalone test.
        # max_det    : 1000 covers dense canopy scenes.
        'yolo_imgsz':      640,
        'resolution':      1.0,
        'conf_threshold':  0.10,
        'iou_threshold':   0.20,
        'max_det':         1000,

        # ── SAM 2 segmentation ────────────────────────────────────────────────
        'sam_checkpoint':  '/content/drive/MyDrive/AGRI/TreeCrown_Segmentation/models/segmentation/sam2.1_t.pt',
        'sam2_batch_size': 4,
        'sam2_max_crop':   640,
        'sam2_min_crop':   32,
        'sam2_sort_grid':  800,
        'sam2_crop_pad':   80,

        # ── Visualisation & output ────────────────────────────────────────────
        'max_viz_dimension': 4000,
        'viz_strip_height':  512,
        'output_dir':  '/content/drive/MyDrive/AGRI/TreeCrown_Segmentation/output',
        'image_name':  'screenshot_result',
    }

    from ultralytics import YOLO
    yolo_model = YOLO(
        '/content/drive/MyDrive/AGRI/TreeCrown_Segmentation/models/detection/yolo12n-150epoch-v1.pt'
    )

    state = DetectionState(
        image_path=SCREENSHOT_PATH,
        model=yolo_model,
        config=config,
    )
    pipeline = create_detection_pipeline(config)
    final    = pipeline.run(state)
    final.print_performance_summary()
```
